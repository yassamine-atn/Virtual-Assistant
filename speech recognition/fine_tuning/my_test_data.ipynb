## Removing special caracters

```python
import re
chars_to_ignore_regex = '[\,\?\.\!\-\;\:\"]'
x=range(len(data))
for i in x : 
  data["text"][i] = re.sub(chars_to_ignore_regex, '', data["text"][i]).lower() + " "
 
```
